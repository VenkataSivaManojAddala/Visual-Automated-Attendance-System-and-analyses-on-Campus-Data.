```python
import pandas as pd
import matplotlib.pyplot as plt


subvsatt_path = "subvsatt.csv"

df = pd.read_csv(subvsatt_path)

a = df.plot(kind='bar', x='PAPER_ID', y='AVG_ATTENDANCE', legend=True)
for p in a.patches:
    a.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
a.set_xticklabels(a.get_xticklabels(), rotation=0)
plt.title('Subject VS Average Attendance')
plt.xlabel('PAPER_ID')
plt.ylabel('AVG_ATTENDANCE')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = 'PLOT.csv'
data = pd.read_csv(file_path)

plt.figure(figsize=(12, 6))

bar_width = 0.35

bar_positions_count = np.arange(len(data['DEPARTMENT_ADMISSION']))
bar_positions_total_students = bar_positions_count + bar_width

plt.bar(bar_positions_count, data['COUNT'], width=bar_width, color='blue', label='No. of students who scored <= 80%', alpha=0.7)
plt.bar(bar_positions_total_students, data['TOTAL_STUDENTS'], width=bar_width, color='orange', label='Total Students Admitted', alpha=0.7)

plt.xlabel('DEPARTMENT_ID')
plt.ylabel('No. of Students')
plt.title('DEPT VS NUMBER OF STUDENTS VS NUMBER OF LOW PERFORMERS IN 2014(SEM:05)')

plt.legend()

# Annotate the values on top of the bars
for i, count in enumerate(data['COUNT']):
    plt.text(bar_positions_count[i] + bar_width / 2, count + 0.1, str(count), ha='right', va='bottom', color='blue', fontsize = 6)

for i, total_students in enumerate(data['TOTAL_STUDENTS']):
    plt.text(bar_positions_total_students[i] + bar_width / 2, total_students + 0.1, str(total_students), ha='right', va='bottom', color='orange', fontsize = 6)

plt.xticks(bar_positions_count + bar_width / 2, data['DEPARTMENT_ADMISSION'], rotation=45, ha='right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'distreasons.csv'
data = pd.read_csv(file_path)

reasons = data['REASON']
num_students = data['NO. OF STUDENTS']


plt.figure(figsize=(8, 8))
plt.pie(num_students, labels=reasons, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors, wedgeprops=dict(width=0.25))


center_circle = plt.Circle((0, 0), 0.6, color='white', fc='white', linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(center_circle)


plt.title('Distribution of different type of Reasons')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your CSV file is named 'data.csv'
file_path = 'SCPLOT.csv'
data = pd.read_csv(file_path)

# Extracting data from the DataFrame
years = data['YEAR']
count = data['COUNT']

# Scatter plot with a line
plt.figure(figsize=(10, 6))
plt.scatter(years, count, color='blue', label='Data Points', marker='o')
plt.plot(years, count, color='orange', linestyle='-', linewidth=2, label='Line of Best Fit')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Scatter Plot with Line')

# Adding grid for better readability
plt.grid(True)

# Adding legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your CSV file is named 'SCPLOT.csv'
file_path = 'SCPLOT.csv'
data = pd.read_csv(file_path)

# Extracting data from the DataFrame
years = data['YEAR']
count = data['COUNT']

# Neat and stylish horizontal bar plot
plt.figure(figsize=(8, 5))
bars = plt.barh(years, count, color='lightblue', edgecolor='black', height=0.6, linewidth=0.8)

# Adding labels and title
plt.xlabel('No. of students joined in DEPT2425', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.title('Horizontal Bar Plot', fontsize=14)

# Set x-axis ticks to integer values only
plt.locator_params(axis='x', integer=True)

# Add annotations to each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.0f}', 
             va='center', ha='left', fontsize=10, color='black')

# Show the plot
plt.show()
```
